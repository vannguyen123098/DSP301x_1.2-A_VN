# bài 1 - Ex1.py

Dòng code `data = re.findall(r'^>.*', str, re.MULTILINE)` sử dụng module `re` trong Python để tìm kiếm tất cả các chuỗi trong biến `str` mà bắt đầu bằng ký tự ">".

- `r'^>.*'`: Đây là biểu thức chính quy (regular expression).
  - `^` ở đầu biểu thức kí hiệu cho việc kiểm tra xem chuỗi bắt đầu từ vị trí nào.
  - `>` là ký tự ">" cần phải xuất hiện đầu tiên.
  - `.*` kí hiệu cho việc lấy bất kỳ ký tự nào (`.`) và có thể xuất hiện 0 hoặc nhiều lần (`*`).

- `str`: Là chuỗi nguồn mà bạn đang tìm kiếm.

- `re.MULTILINE`: Khi sử dụng cờ này, dấu `^` sẽ áp dụng cho mỗi dòng trong chuỗi, không chỉ áp dụng cho chuỗi trong toàn bộ nội dung.

Do đó, dòng code này sẽ trả về một danh sách `data` chứa tất cả các dòng trong chuỗi `str` mà bắt đầu bằng ký tự ">".

```python
import re

str="""
>Venues
>Marketing
>medalists
>Controversies
>Paralympics
>snowboarding
>[1]
>Netherlands
>[2]
>Norway
>[10]
>[11]
>References
>edit
>[12]
>Norway
>Germany
>Canada
>Netherlands
>Japan
>Italy
>Belarus
>China
>Slovakia
<$#%#$%
<#$#$$
<**&&^^
>Slovenia
>Belgium
>Spain
>Kazakhstan
>[15]
>1964
>1968
>1972
>1992
>1996
>2000"""

#Type your answer here.

data = re.findall(r'^>.*', str, re.MULTILINE)

print(data)
```

# Bài 2:
Mở file Ex2.py và hoàn thành đoạn code để lấy được các email có trong chuỗi đó. Kết quả sau khi chạy sẽ như sau:

['franky@google.com', 'sinatra123@yahoo.com']

Trong biểu thức chính quy `r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'`:

- `\b`: Ký hiệu cho biên giới từ.
- `[A-Za-z0-9._%+-]+`: Địa chỉ email có thể chứa bất kỳ chữ cái, số, hoặc các ký tự đặc biệt như ".", "_", "%", "+", "-".
- `@`: Ký tự "@" phải xuất hiện.
- `[A-Za-z0-9.-]+`: Tên miền (domain) của email có thể chứa bất kỳ chữ cái, số, hoặc các ký tự đặc biệt như ".", "-".
- `\.`: Dấu chấm trước đuôi domain.
- `[A-Z|a-z]{2,}`: Phần đuôi domain có thể chứa ít nhất 2 chữ cái (ví dụ: com, org, vn, ...).
- `\b`: Ký hiệu cho biên giới từ.

Kết quả sau khi chạy sẽ là danh sách các địa chỉ email trong chuỗi.

```python
import re

str='The advancements in biomarine studies franky@google.com with the investments necessary and Davos sinatra123@yahoo.com Then The New Yorker article on wind farms...'
#Type your answer here.

regex= r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
emails=re.findall(regex, str)


print(emails)
```

# Bài 3:

Mở file Ex3.py và hoàn thành đoạn code để lấy được các email có trong chuỗi đó (chỉ lấy phần ở trước ký tự '@' và ký tự '@'). Kết quả sau khi chạy sẽ như sau:

['franky@', 'sinatra123@']

Trong biểu thức chính quy `r'(\S+)@'`:

- `(\S+)`: Là một nhóm (group) trong đó `\S` đại diện cho bất kỳ ký tự không phải là khoảng trắng và `+` đại diện cho việc có ít nhất một ký tự không phải khoảng trắng.
- `@`: Là ký tự '@' cần phải xuất hiện.

Kết quả sau khi chạy sẽ là danh sách các phần trước ký tự '@' trong chuỗi.

```python
import re

str='The advancements in biomarine studies franky@google.com, with the investments necessary and Davos sinatra123@yahoo.com Then The New Yorker article on wind farms...'
#Type your answer here.

#regex= r'\b[A-Za-z0-9._%+-]+@'
regex = r'(\S+@)'
emails=re.findall(regex, str)


print(emails)
```

# Bài 4:

Mở file Ex4.py và hoàn thành đoạn code tương tự như bài 3, nhưng lần này không lấy ký tự '@' nữa. Kết quả sau khi chạy sẽ như sau:

['franky', 'sinatra123']

Trong biểu thức chính quy r'(\S+)(?=@)':

- (\S+): Là một nhóm (group) trong đó \S đại diện cho bất kỳ ký tự không phải là khoảng trắng và + đại diện cho việc có ít nhất một ký tự không phải khoảng trắng.

- (?=@): Đây là một positive lookahead assertion, nó đảm bảo rằng ký tự '@' không được bao gồm trong kết quả.
Kết quả sau khi chạy sẽ là danh sách các phần trước ký tự '@' trong chuỗi.

```python
import re

str='The advancements in biomarine studies franky@google.com, with the investments necessary and Davos sinatra123@yahoo.com Then The New Yorker article on wind farms...'
#Type your answer here.

#regex= r'(\S+)(?=@)'
regex = r'(\S+)@'
emails=re.findall(regex, str)


print(emails)
```
