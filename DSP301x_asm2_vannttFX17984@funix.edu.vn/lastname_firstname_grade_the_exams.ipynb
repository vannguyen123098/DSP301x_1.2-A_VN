```python
# Task 3
def calculate_score(student_answers, answer_key):
    score = 0
    correct_answers = answer_key.split(',')
    for student_ans, correct_ans in zip(student_answers, correct_answers):
        if student_ans == correct_ans:
            score += 4
        elif student_ans:
            score -= 1
    return max(0, score)

def count_high_scores(scores):
    return sum(score > 80 for score in scores)

def calculate_mean(scores):
    return round(sum(scores) / len(scores), 3) if scores else 0

def calculate_range(scores):
    return max(scores) - min(scores) if scores else 0

def calculate_median(scores):
    sorted_scores = sorted(scores)
    n = len(sorted_scores)
    if n % 2 == 0:
        return round((sorted_scores[n // 2 - 1] + sorted_scores[n // 2]) / 2, 3)
    else:
        return sorted_scores[n // 2]

def print_skipped_questions(skipped_questions, total_students):
    print('Question that most people skip: ')
    for question, count in sorted(skipped_questions.items(), key=lambda x: x[1], reverse=True):
        print(f'{question} - {count} - {count / total_students:.2%}', end=', ')
    print()

def print_incorrect_questions(incorrect_questions, total_students):
    print('Question that most people answer incorrectly: ')
    for question, count in sorted(incorrect_questions.items(), key=lambda x: x[1], reverse=True):
        print(f'{question} - {count} - {count / total_students:.2%}', end=', ')
    print()

def grade_exam(file_name, answer_key):
    valid_lines = 0
    invalid_lines = 0
    scores = []
    skipped_questions = {}
    incorrect_questions = {}

    with open(f'/content/drive/MyDrive/funix_DS/DSP301x_1.2-A_VN Giới thiệu về Khoa học Dữ liệu/DSP301x_asm2_vannttFX17984@funix.edu.vn/Data Files/{file_name}.txt', 'r') as file:
        print(f'Successfully opened {file_name}')
        print('**** ANALYZING ****')

        for line_number, line in enumerate(file, 1):
            values = line.strip().split(',')

            if len(values) == 26:
                student_id = values[0]
                if student_id[0] == 'N' and student_id[1:].isdigit():
                    valid_lines += 1
                    student_answers = values[1:]
                    current_score = calculate_score(student_answers, answer_key)
                    scores.append(current_score)

                    for i, answer in enumerate(student_answers, 1):
                        if not answer:
                            skipped_questions[i] = skipped_questions.get(i, 0) + 1

                    for i, (student_ans, correct_ans) in enumerate(zip(student_answers, answer_key.split(',')), 1):
                        if student_ans and student_ans != correct_ans:
                            incorrect_questions[i] = incorrect_questions.get(i, 0) + 1

                else:
                    print(f'Invalid line of data (line {line_number}): #N is invalid')
                    invalid_lines += 1
            else:
                print(f'Invalid line of data (line {line_number}): does not contain exactly 26 values')
                invalid_lines += 1

    total_students = valid_lines - invalid_lines
    print('**** REPORT ****')
    print(f'Total valid lines of data: {valid_lines}')
    print(f'Total invalid lines of data: {invalid_lines}')
    print(f'Total student of high scores: {count_high_scores(scores)}')
    print(f'Mean (average) score: {calculate_mean(scores)}')
    print(f'Highest score: {max(scores)}')
    print(f'Lowest score: {min(scores)}')
    print(f'Range of scores: {calculate_range(scores)}')
    print(f'Median score: {calculate_median(scores)}')
    print_skipped_questions(skipped_questions, total_students)
    print_incorrect_questions(incorrect_questions, total_students)

def main():
    answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
    while True:
        file_name = input('Enter a class to grade (i.e. class1 for class1.txt): ')
        if file_name.lower() == 'quit':
            print('Exiting the program')
            break

        grade_exam(file_name, answer_key)

if __name__ == '__main__':
    main()
```

```python
#task 5

import pandas as pd
import numpy as np

def read_file(file_name):
    try:
        filePath = f'/content/drive/MyDrive/funix_DS/DSP301x_1.2-A_VN Giới thiệu về Khoa học Dữ liệu/DSP301x_asm2_vannttFX17984@funix.edu.vn/Data Files/{file_name}.txt'
        with open(filePath, 'r') as file:
            print(f"Successfully opened {file_name}.txt")
            lines =  file.readlines()
            print('File Content:')
            print(''.join(lines))
            return lines

    except FileNotFoundError:
        print(f"File {file_name}.txt not found.")
        return None

def analyze_data(lines, answer_key):
    if lines is None:
        return

    valid_lines = 0
    invalid_lines = 0
    scores = []

    # Initialize dictionaries to track skipped and incorrect answers for each question
    skipped_questions = {}
    incorrect_questions = {}

    for line in lines:
        data = line.strip().split(',')

        if len(data) == 26 and data[0].startswith('N') and data[0][1:].isdigit():
            valid_lines += 1
            student_id = data[0]
            student_answers = data[1:]

            score = 0
            for ans, key in zip(student_answers, answer_key.split(',')):
                if ans == '':
                    # Skip, no points deducted or awarded
                    skipped_questions[key] = skipped_questions.get(key, 0) + 1
                elif ans == key:
                    score += 4
                else:
                    score -= 1
                    incorrect_questions[key] = incorrect_questions.get(key, 0) + 1

            scores.append((student_id, score))
        else:
            print(f'Invalid line of data: {line}')
            invalid_lines += 1

    print('**** REPORT ****')
    print(f'Total valid lines of data: {valid_lines}')
    print(f'Total invalid lines of data: {invalid_lines}')

    # Calculate statistics
    high_scores = sum(score > 80 for _, score in scores)
    mean_score = round(np.mean([score for _, score in scores]), 2)
    highest_score = max([score for _, score in scores])
    lowest_score = min([score for _, score in scores])
    range_of_scores = highest_score - lowest_score
    median_score = np.median([score for _, score in scores])

    print(f'Total student of high scores: {high_scores}')
    print(f'Mean (average) score: {mean_score}')
    print(f'Highest score: {highest_score}')
    print(f'Lowest score: {lowest_score}')
    print(f'Range of scores: {range_of_scores}')
    print(f'Median score: {median_score}')

    # Sort and find most skipped questions
    most_skipped = sorted(skipped_questions.items(), key=lambda x: x[1], reverse=True)
    print('Question that most people skip:', ', '.join([f'{q} - {skipped} - {round(skipped / valid_lines *100, 2)}%' for q, skipped in most_skipped]))

    # Sort and find most incorrectly answered questions
    most_incorrect = sorted(incorrect_questions.items(), key=lambda x: x[1], reverse=True)
    print('Question that most people answer incorrectly:', ', '.join([f'{q} - {incorrect} - {round(incorrect / valid_lines *100, 2)}%' for q, incorrect in most_incorrect]))

    return scores

def write_grades_to_file(file_name, scores):
    output_file_name = file_name.replace('.txt', '_grades.txt')
    with open(output_file_name, 'w') as output_file:
        for student_id, score in scores:
            output_file.write(f'{student_id},{score}\n')
    print(f'Grades written to {output_file_name}')

def main():
    while True:
        file_name = input("Enter a class file to grade (e.g., class1 for class1.txt): ")
        if file_name.lower() == 'quit':
            print('Exiting the program')
            break

        if not file_name:
            print("file_name cannot be empty.")
            continue

        lines = read_file(file_name)

        if lines is not None:
            print('**** ANALYZING ****')
            answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
            scores = analyze_data(lines, answer_key)
            write_grades_to_file(file_name, scores)

if __name__ == "__main__":
    main()
```
